**Introduction**

This project explores these differences by analyzing how both humans and AI respond to various real-life scenarios. By comparing their responses, we aim to understand patterns in their thinking, such as logical reasoning, emotional influence, ethical considerations, and risk-taking behavior.

Understanding these differences is important for evaluating the strengths and limitations of AI in decision-making tasks and its role alongside human judgment.

**Problem Statement**

Although AI systems are increasingly used for decision-making, there is still a lack of clear understanding of how their decision-making behavior differs from that of humans.

The problem addressed in this study is to identify and analyze these differences in decision-making patterns between AI and humans, and to understand how each responds under similar conditions.

## Step 3.1: Statistical Comparison using T-Test (Human vs AI)

Note:
The response categories were numerically encoded to enable comparative statistical analysis.
Although the original data is categorical, numerical encoding was used as an approximation for measuring distributional differences between Human and AI responses.

```python
import pandas as pd
from scipy import stats

# Set display options to show all rows
pd.set_option('display.max_rows', None)

t_test_results = []
question_cols = [f'Q{i+1}' for i in range(31)]

for col in question_cols:
    human_grp = df_final[df_final['Source'] == 0][col]
    ai_grp = df_final[df_final['Source'] == 1][col]

    t_stat, p_val = stats.ttest_ind(
        human_grp,
        ai_grp,
        equal_var=False,
        nan_policy='omit'
    )

    t_test_results.append({
        'Question': col,
        'T_Statistic': t_stat,
        'T_P_Value': p_val
    })

df_t_test = pd.DataFrame(t_test_results)

print("📊 3.1: T-Test Results (Full Dataset)")
display(df_t_test)
```

In this step, statistical hypothesis testing is performed to determine whether there is a significant difference between Human and AI responses for each question.

1. All question columns (Q1 to Q31) were selected for analysis.

2. For each question:
   - Data was divided into two groups:
     - Human responses (Source = 0)
     - AI responses (Source = 1)

3. An **Independent T-Test (Welch’s T-Test)** was applied:
   - This test compares the mean of two independent groups

4. Two values were calculated for each question:
   - **T-Statistic** → measures difference between groups  
   - **P-Value** → tells whether the difference is statistically significant  

5. All results were stored in a structured DataFrame for further analysis.

###  Result:
The output shows, for each question, whether Human and AI responses differ significantly. Questions with low p-values indicate strong differences in decision-making patterns.

##  Step 3.3: Effect Size Analysis using Cohen’s d



Note:
Cohen’s d values are interpreted as approximate effect sizes based on encoded categorical responses rather than true continuous measurements.

```python
import numpy as np

# Improved function with a safety check for zero variance
def calculate_cohen_d(g1, g2):
    n1, n2 = len(g1), len(g2)
    v1, v2 = np.var(g1, ddof=1), np.var(g2, ddof=1)
    pooled_sd = np.sqrt(((n1 - 1) * v1 + (n2 - 1) * v2) / (n1 + n2 - 2))
    # If pooled_sd is 0, it means both groups answered exactly the same
    return (np.mean(g1) - np.mean(g2)) / pooled_sd if pooled_sd > 0 else 0

cohen_results = []

for col in question_cols:
    h = df_final[df_final['Source'] == 0][col]
    a = df_final[df_final['Source'] == 1][col]

    # Calculate absolute difference magnitude
    d = abs(calculate_cohen_d(h, a))

    # Storing as a dictionary for better tracking
    cohen_results.append({'Question': col, 'Cohen_d': d})

# Creating a labeled DataFrame
df_cohen = pd.DataFrame(cohen_results)

print("✅ 3.3: Labeled Cohen's d Results Ready")
display(df_cohen)
```

In this step, Cohen’s d is calculated to measure the magnitude of difference between Human and AI responses for each question.

### Purpose of this Step:
- To quantify how strong the difference is between Human and AI responses
- To complement statistical significance tests (T-test and Chi-Square)
- To understand practical importance of differences, not just statistical significance

###  What was done:

1. For each question (Q1 to Q31), responses were divided into two groups:
   - Human responses (Source = 0)
   - AI responses (Source = 1)

2. The mean and variance of both groups were calculated.

3. A pooled standard deviation was computed to normalize the difference.

4. Cohen’s d formula was applied:
   - Measures difference between group means relative to variability

5. A safety check was added:
   - If both groups have zero variance (identical responses), effect size is set to 0 to avoid division errors

6. Absolute values were used:
   - Focus is on magnitude of difference, not direction

7. Results were stored in a labeled DataFrame (`df_cohen`) for clear interpretation.

###  Result:
The output provides effect size values for each question, indicating how strongly Human and AI responses differ in decision-making behavior.

```python
import numpy as np
import pandas as pd
from scipy import stats
from statsmodels.stats.multitest import multipletests

# ---------------------------------------------------------
# AUTO-FIX: RE-CALCULATING MISSING VALUES
# ---------------------------------------------------------
question_cols = [f'Q{i+1}' for i in range(31)]

# 1. Check if T-test data exists, if not, calculate it
if 'df_t_test' not in locals():
    print("🔄 Recovering T-test data...")
    t_results = []
    for col in question_cols:
        h = df_final[df_final['Source'] == 0][col]
        a = df_final[df_final['Source'] == 1][col]
        t_stat, p_val = stats.ttest_ind(h, a, nan_policy='omit')
        t_results.append({'Question': col, 'T_Statistic': t_stat, 'T_P_Value': p_val})
    df_t_test = pd.DataFrame(t_results)

# 2. Check if Chi-square data exists
if 'df_chi' not in locals():
    print("🔄 Recovering Chi-square data...")
    c_results = []
    for col in question_cols:
        contingency = pd.crosstab(df_final['Source'], df_final[col])
        chi2, p_chi, _, _ = stats.chi2_contingency(contingency)
        c_results.append({'Question': col, 'Chi_P_Value': p_chi})
    df_chi = pd.DataFrame(c_results)

# 3. Check if Cohen's d list exists (Fixes your specific NameError)
if 'cohen_list' not in locals():
    print("🔄 Recovering Cohen's d effect sizes...")
    def calc_d(g1, g2):
        n1, n2 = len(g1), len(g2)
        v1, v2 = np.var(g1, ddof=1), np.var(g2, ddof=1)
        sd = np.sqrt(((n1-1)*v1 + (n2-1)*v2) / (n1+n2-2))
        return (np.mean(g1) - np.mean(g2)) / sd if sd > 0 else 0
    cohen_list = [abs(calc_d(df_final[df_final['Source']==0][c], df_final[df_final['Source']==1][c])) for c in question_cols]

# ---------------------------------------------------------
# FINAL TABLE ASSEMBLY
# ---------------------------------------------------------
df_inference = df_t_test.copy()
df_inference['Chi_P_Value'] = df_chi['Chi_P_Value']
df_inference['Cohen_d'] = cohen_list

# Bonferroni Correction for strict accuracy
p_adj = multipletests(df_inference['Chi_P_Value'], method='bonferroni')[1]
df_inference['Adjusted_P_Value'] = p_adj
df_inference['Significant'] = df_inference['Adjusted_P_Value'] < 0.05

# Sorting by impact
df_inference = df_inference.sort_values(by='Cohen_d', ascending=False)

print("\n🏆 FINAL STATISTICAL INFERENCE TABLE READY")
display(df_inference)
```

```python
import plotly.graph_objects as go

# Use the dataframe that contains original text categories (from Step 2)
human_data = df_viz[df_viz['Source'] == 'Human']
ai_data = df_viz[df_viz['Source'] == 'AI']

# Get all unique decision categories from the data
categories = []
for col in question_cols:
    categories.extend(human_data[col].unique())
    categories.extend(ai_data[col].unique())
categories = sorted(set(categories))

# Remove 'unknown' if it exists (not needed for radar chart)
categories = [c for c in categories if c != 'unknown']

# Function to calculate percentage for each category
def get_percentages(data, categories, question_cols):
    counts = []
    total = 0
    for cat in categories:
        # Count how many times this category appears in all answers
        count = (data[question_cols] == cat).sum().sum()
        counts.append(count)
        total += count
    # Convert counts to percentages
    percentages = [c / total * 100 for c in counts]
    return percentages

# Calculate percentages for humans and AI
human_percentages = get_percentages(human_data, categories, question_cols)
ai_percentages = get_percentages(ai_data, categories, question_cols)

# Create radar chart
fig = go.Figure()
fig.add_trace(go.Scatterpolar(
    r=human_percentages,
    theta=categories,
    fill='toself',
    name='Human Decisions',
    line_color='blue'
))
fig.add_trace(go.Scatterpolar(
    r=ai_percentages,
    theta=categories,
    fill='toself',
    name='AI Decisions',
    line_color='red'
))
fig.update_layout(
    polar=dict(radialaxis=dict(visible=True, range=[0, 100])),
    showlegend=True,
    title="Decision Patterns: Human vs AI (Percentage Frequency)"
)
fig.show()
```

 Phase 6.2: Hierarchical Clustering using Cosine Distance

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.metrics.pairwise import cosine_distances

# =========================================================
# PHASE 6.2: HIERARCHICAL CLUSTERING USING COSINE DISTANCE
# =========================================================

# STEP 1: AUTOMATIC DATAFRAME DETECTION

for var_name in ['df_viz', 'df_ml', 'df_final', 'df_combined', 'df']:
    if var_name in locals():
        df_active = locals()[var_name].copy()
        break

# Extract all question columns automatically
questions = [c for c in df_active.columns if c.startswith('Q') and c[1:].isdigit()]

# Detect all encoded response values present in the dataset
all_possible_values = sorted(pd.unique(df_active[questions].values.ravel()))

# =========================================================
# STEP 2: BUILD RESPONSE DISTRIBUTION FEATURE MATRIX
# =========================================================

feature_matrix = []

for q in questions:

    # Calculate normalized response frequencies
    counts = df_active[q].value_counts(normalize=True)

    # Create distribution vector for each question
    vec = [counts.get(val, 0.0) for val in all_possible_values]

    feature_matrix.append(vec)

feature_matrix = np.array(feature_matrix)

# =========================================================
# STEP 3: COMPUTE COSINE DISTANCE MATRIX
# =========================================================

distance_matrix = cosine_distances(feature_matrix)

# =========================================================
# STEP 4: HIERARCHICAL CLUSTERING
# =========================================================

linkage_matrix = linkage(distance_matrix, method='ward')

# =========================================================
# STEP 5: VISUALIZE DENDROGRAM
# =========================================================

plt.figure(figsize=(12, 6))

dendrogram(
    linkage_matrix,
    labels=questions,
    leaf_rotation=90,
    leaf_font_size=10
)

plt.title(
    'Question Clustering based on Decision Patterns (Cosine Distance)',
    fontsize=13,
    fontweight='bold',
    pad=15
)

plt.xlabel(
    'Questions',
    fontsize=11
)

plt.ylabel(
    'Cluster Distance',
    fontsize=11
)

plt.tight_layout()
plt.show()

print("Success: Hierarchical clustering completed using Cosine Distance.")
```

### **Phase 6: Unsupervised Behavioral Clustering via Hierarchical Dendrogram Analysis**

#### **1. Objective & Methodology**
Instead of using predictive machine learning models, this analysis applies an **Unsupervised Learning approach**—specifically **Hierarchical Agglomerative Clustering**—to explore how the 31 decision scenarios group together naturally.

The goal is to understand which questions share similar response patterns without using any manual labels, predefined categories, or human bias.

The automated process works through the following steps:
1. **Dynamic Vectorization:** The code automatically calculates the percentage distribution of the encoded responses for each question to build its behavioral profile.
2. **Cosine Distance Metric:** Instead of measuring absolute numbers, we use **Cosine Distance** to find the geometric similarity between questions. This focuses entirely on the overall pattern and shape of the data distribution.
3. **Ward’s Linkage Criterion:** The tree is built using **Ward’s Method**, which connects the most similar questions first while keeping the variation within each group as low as possible.

---

#### **2. Automated Process Breakdown**

* **Automatic Option Detection:** The script automatically scans the dataset to detect every encoded category present across the columns. This keeps the code 100% data-driven and removes any risk of manual typing errors.
* **Proportional Analysis:** It evaluates the exact ratio of choices selected for each scenario, turning every question into a clear statistical pattern.
* **Dendrogram Visualization:** The final clustering tree connects individual questions into small branches, which then merge into larger groups based on their statistical similarity.

---

#### **3. Core Insights & Interpretations**

* **Close Connections (Short Vertical Distances):** Questions that link together very quickly at the bottom of the graph share nearly identical response patterns. This means participants answered these specific real-world scenarios using a highly consistent decision style.
* **Higher-Level Clusters (Main Groupings):** The clear separation of the tree into distinct colored branches shows fundamental differences in how scenarios were perceived. For example, highly emotional choices naturally split away into separate branches compared to scenarios based on routine logic or risk assessment.
* **Validation of the Survey Design:** The fact that the algorithm creates clean, well-grouped clusters instead of scattering the questions randomly proves that the survey instrument is highly stable and successful in capturing distinct behavioral patterns.
